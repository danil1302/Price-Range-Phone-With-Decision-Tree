## Apply Model

```python
X = data.drop(['price_range'], axis=1)

y = data['price_range']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state=42)
```

## Training

### Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
klf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)

klf_gini.fit(X_train, y_train)
```

```python
y_pred_gini = klf_gini.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Skor akurasi model dengan kriteria index gini: {0:0.4f}'. format(accuracy_score(y_test,y_pred_gini)))
```

```python
y_pred_train_gini = klf_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Training-set akurasi skor : {0:0.4f}'.format(accuracy_score(y_train, y_pred_train_gini)))
```

## Check Overfitting dan Underfitting

```python
print('Training set skor: {:.4f}'.format(klf_gini.score(X_train, y_train)))

print('Test set skor: {:.4f}'.format(klf_gini.score(X_test, y_test)))
```

## Visualisasi Decision Tree

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(klf_gini.fit(X_train, y_train))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,15))
plot_tree(klf_gini, max_depth = 1,  
            rounded = True,
            filled = True, 
            impurity = True, 
            node_ids = True,
          )
plt.show()
```

## Classifier with criterion entropy

```python
klf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)


klf_en.fit(X_train, y_train)
```

```python
y_pred_en = klf_en.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Skor model akurasi dengan criterion entropy: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

```python
y_pred_train_en = klf_en.predict(X_train)

y_pred_train_en
```

```python
print('Training-set akurasi skor: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

con_mat = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', con_mat)
```

## Confusion Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_en))
```
